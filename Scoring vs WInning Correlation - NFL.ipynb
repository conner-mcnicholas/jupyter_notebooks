```python
from sportsipy.nfl.teams import Teams
from scipy.stats import pearsonr
import pandas as pd

print('Correlation(WinRatio,GoalDifferential)\n')

for year in range(1980, 2021):
    teams = list(Teams(year))
    statd = {}
    for i,team in enumerate(teams[0:len(teams)]):
        statd[i] = {#'TEA' : team.abbreviation,
                    'Team' : team.name,
                    #'Rank' : team.rank,
                    'W' : team.wins,
                    'L' : team.losses,
                    'GP' : team.games_played,
                    'Rwin' : round(team.wins/team.games_played,4),
                    'GF' : team.points_for,
                    'GA' : team.points_against,
                    'GD' : team.points_for-team.points_against,
                    'GD_per' : (team.points_for-team.points_against)/team.games_played}

    df = pd.DataFrame(statd)
    df = df.transpose()
    
    x = list(df['Rwin'])
    seq = sorted(x,reverse=True)
    index = [seq.index(v) for v in x]
    df['Iwin'] =  index
  
    x = list(df['GD'])
    seq = sorted(x,reverse=True)
    index = [seq.index(v) for v in x]
    df['Idiff'] =  index
    
    valuCorr, _ = pearsonr(df['GD'].tolist(), df['Rwin'].tolist())
    print('%s: %s' % (year,round(valuCorr,3)))
```

```python
from sportsipy.nfl.teams import Teams
from scipy.stats import pearsonr
import pandas as pd
    
year=2004
teams = list(Teams(year))
statd = {}
    
for i,team in enumerate(teams[0:len(teams)]):
    statd[i] = {'Team' : team.name,
                'W' : team.wins,
                'L' : team.losses,
                'G' : team.games_played,
                'Rwin' : round(team.wins/team.games_played,3),
                'GF' : team.points_for,
                'GA' : team.points_against,
                'GD' : team.points_for-team.points_against,
                'GD_per' : (team.points_for-team.points_against)/team.games_played}

df = pd.DataFrame(statd)
df = df.transpose()

x = list(df['Rwin'])
seq = sorted(x,reverse=True)
index = [seq.index(v) for v in x]
df['Iwin'] =  index

x = list(df['GD'])
seq = sorted(x,reverse=True)
index = [seq.index(v) for v in x]
df['Idiff'] =  index

df = df[['Team','W','L','G','GF','GA','GD','Rwin','GD_per','Iwin','Idiff']]
df = df.sort_values('Rwin',ascending=False)
print('%s MLB Season - Ranked by [Iwin] -> Ratio(Games Won:Total Games) [Rwin] \n' % year)
print(df.to_string(index=False))

df = df[['Team','W','L','G','GF','GA','GD','Rwin','GD_per','Iwin','Idiff']]
df = df.sort_values('GD',ascending=False)
print('\n%s MLB Season - Ranked by [Idiff] -> Goal Differential [GD] \n' % year)
print(df.to_string(index=False))
```
